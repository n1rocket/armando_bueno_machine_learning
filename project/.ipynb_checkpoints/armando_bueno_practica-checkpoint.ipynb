```python
# Creo una función para realizar la limpieza de campos que no se consideran necesarios
def clean_dataframe(dataframe):
    result = dataframe
    # Limpio el dataset de descripciones introducidas por personas, identificadores de tablas, urls e información no relevante
    result = result.drop(['ID','Listing Url','Scrape ID','Last Scraped',
                          'Name','Summary','Space','Description','Experiences Offered',
                         'Neighborhood Overview','Notes','Transit','Access','Interaction',
                         'House Rules','Thumbnail Url','Medium Url','Picture Url',
                         'XL Picture Url','Host ID','Host URL','Host Name','Host Location',
                         'Host About','Host Thumbnail Url','Host Picture Url',
                         'Host Neighbourhood','Street',

                          # Revisar "Neighbourhood Cleansed" se puede codificar porque el barrio seguramente afecte al precio,
                          #  utilizaré el Cleansed ya que parece que es un campo "limpiado"
                          'Neighbourhood',
                          # Elimino el grupo de vecindario ya que es más abstracto que el Cleansed y está correlacionado
                          'Neighbourhood Group Cleansed',
                          'City','State','Market','Smart Location','Country Code',
                          'Country','Latitude','Longitude','Geolocation',
                          'Reviews per Month','Jurisdiction Names',
                          'License','First Review','Last Review',
                          # * Comprobar si puedo separar las verificaciones para ver si afecta "Host Verifications"    
                          'Host Verifications',
                          # * Comprobar si puedo separar las Features para ver si afecta "Features"    
                          'Features',
                          # * Comprobar si puedo separar las Features para ver si afecta "Amenities"    
                          'Amenities',
                          # Precios, actualizaciones, disponibilidad, fechas de scrapeo
                          'Weekly Price','Monthly Price','Security Deposit','Calendar Updated','Has Availability',
                          'Availability 30','Availability 60','Availability 90','Availability 365','Calendar last Scraped',
                          'Calculated host listings count','Host Total Listings Count',
                          'Host Acceptance Rate','Host Since',
                         ],axis=1)


    
    
    return result
```

```python
## Columnas que podemos codificar

# Host Response Time
# Property Type
# Room Type
# Cancellation Policy
# Bed Type
# Neighbourhood Cleansed

# Otras opciones: onehot, ordinal, mean... 

from sklearn.preprocessing import LabelEncoder

le_hrt = LabelEncoder()
le_pt = LabelEncoder()
le_rt = LabelEncoder()
le_cp = LabelEncoder()
le_bt = LabelEncoder()
le_nc = LabelEncoder()
le_nc = LabelEncoder()

def encodeColumns(df):
    df['Host Response Time'] = le_hrt.fit_transform(df['Host Response Time'])
    df['Property Type'] = le_pt.fit_transform(df['Property Type'])
    df['Room Type'] = le_rt.fit_transform(df['Room Type'])
    df['Cancellation Policy'] = le_cp.fit_transform(df['Cancellation Policy'])
    df['Bed Type'] = le_bt.fit_transform(df['Bed Type'])
    df['Neighbourhood Cleansed'] = le_nc.fit_transform(df['Neighbourhood Cleansed'])
```

```python
def showCorrelation(df):
    # Correlación
    import seaborn as sns

    # Compute the correlation matrix
    corr = np.abs(df).corr()

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(12, 10))

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask,vmin = 0.0, vmax=1.0, center=0.5,
                linewidths=.1, cmap="YlGnBu", cbar_kws={"shrink": .8})

    plt.show()

showCorrelation(train_df)
```

## Preprocesamiento

a. Eliminación de variables, mediante selección (random forest/Lasso), alta
correlación, alto porcentaje de missings, o el método que se considere
oportuno.

b. Generación de variables

```python
# Elimino campos
# data =  data.drop(['id','Unnamed: 32'],axis=1)

def cleanPreprocessing(df):
    result = df
    # Voy a eliminar los valores con alta correlación visto en la gráfica superior
    # Bed y Bedrooms con Accommodates -> Elimino este último
    result = result.drop(['Accommodates'],axis=1)

    # Bed con Bedrooms, me parece más específico la cantidad de camas
    result = result.drop(['Beds'],axis=1)

    # Host response Rate con Time igual, me quedo con Time
    result = result.drop(['Host Response Rate'],axis=1)

    # Los Reviews también tienen alta correlación
    result = result.drop(
        ['Review Scores Rating', 
         'Review Scores Accuracy', 
         'Review Scores Checkin',
         'Review Scores Communication',
         'Review Scores Cleanliness',
         'Review Scores Location'],axis=1)
    return result
```

## Modelado

### Separo en 'X' y en 'y', creo una función para simplificar y poder repetir luego con test

```python
def data_split_x_y(df, target_column):
    features = df.columns.drop([target_column])
    x = df[features].values
    y = df[target_column].values
    return x,y
```

```python
X_train, y_train = data_split_x_y(train_df, 'Price')
```

### Cross Validation

#### Prueba 1 Tree

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

# Create the decision tree regressor
dt = DecisionTreeRegressor()

maxDepth = range(1,10)

# Define the parameter grid for the grid search , 'min_samples_leaf': [1, 2, 3, 4]
param_grid = {'max_depth': maxDepth}

# Create the grid search object , scoring='neg_mean_squared_error'
grid_search = GridSearchCV(dt, param_grid, cv=5)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)

scores = grid_search.cv_results_['mean_test_score']

print("Scores: ",scores)

plt.plot(maxDepth,scores,'-o')
plt.xlabel('max_depth',fontsize=16)
plt.ylabel('10-Fold MSE')
#plt.ylim((-1, 0))
plt.show()

# Use the best parameters to make predictions on the test set
#y_pred = grid_search.predict(X_test)

# Print the test set mean squared error
#print("Test set mean squared error: ", mean_squared_error(y_test, y_pred))
```

#### Prueba 2 Tree con min_samples_leaf

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

# Create the decision tree regressor
dt = DecisionTreeRegressor()

maxDepth = range(1,10)

# Define the parameter grid for the grid search
param_grid = {'max_depth': maxDepth , 'min_samples_leaf': [1, 2, 3, 4]}

# Create the grid search object 
grid_search_2 = GridSearchCV(dt, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the data
grid_search_2.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters 2: ", grid_search_2.best_params_)
print("Best score 2: ", grid_search_2.best_score_)
```

#### Prueba 3 

```python
from sklearn.ensemble import GradientBoostingRegressor

Niterations = [500,1000,1500,2000]
learningRate = [0.1,0.05]
# mantenemos max_depth estático: max_depth=3

param_grid = {'n_estimators': Niterations,'learning_rate':learningRate }
grid = GridSearchCV(GradientBoostingRegressor(random_state=0, max_depth=3), param_grid=param_grid, cv = 5, verbose=2)
grid.fit(X_train, y_train)
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))
```

### Evaluación del modelo (Varias pruebas)

#### Prueba 1 - Test

```python
maxDepthOptimo = grid_search.best_params_['max_depth']

treeModel = DecisionTreeRegressor(max_depth=maxDepthOptimo).fit(X_train,y_train)

print("Train: ",treeModel.score(X_train,y_train))
print("Test: ",treeModel.score(X_test,y_test))
```

#### Prueba 2 - Test

```python
maxDepthOptimo2 = grid_search_2.best_params_['max_depth']
minSamplesLeafOptimo2 = grid_search_2.best_params_['min_samples_leaf']

treeModel2 = DecisionTreeRegressor(max_depth=maxDepthOptimo2, min_samples_leaf=minSamplesLeafOptimo2).fit(X_train,y_train)

print("Train: ",treeModel2.score(X_train,y_train))
print("Test: ",treeModel2.score(X_test,y_test))
```

#### Prueba 3 - Test

```python
lrOptimo = grid.best_params_['learning_rate']
neOptimo = grid.best_params_['n_estimators']
bt = GradientBoostingRegressor(random_state=0, max_depth=3,learning_rate=lrOptimo, n_estimators=neOptimo)
bt.fit(X_train,y_train)

error = 1-grid.cv_results_['mean_test_score'].reshape(len(learningRate),len(Niterations))
colors = ['r','b','g','k','m']
for i,lr in enumerate(learningRate):    
    plt.plot(Niterations,error[i,:],colors[i] + '--o',label='lr = %g'%lr)

plt.legend()
plt.xlabel('# iteraciones')
plt.ylabel('5-fold CV Error')
plt.title('train: %0.3f\ntest:  %0.3f'%(bt.score(X_train,y_train),bt.score(X_test,y_test)))
plt.grid()
plt.show()
```

```python
importances = bt.feature_importances_
importances = importances / np.max(importances)

indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,10))
plt.barh(range(X_train.shape[1]),importances[indices])
plt.yticks(range(X_train.shape[1]),features[indices])
plt.show()
```
