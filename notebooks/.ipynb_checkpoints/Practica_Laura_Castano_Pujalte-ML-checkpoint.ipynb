## 1.Leer los datos

##### Utilizamos el dataset de Airbnb filtrado por Madrid con 14780 entradas.

```python
import pandas as pd
#path = 'C:/temp/'
airbnb_extract = pd.read_csv('airbnb-listings-extract.csv', sep= ';', decimal='.')
airbnb_extract
```

#### 1.Preparación de los datos

##### Dividir los datos entre train y test:

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(airbnb_extract, test_size=0.2, shuffle=True, random_state=0)

print(f'Dimensiones del dataset de training: {train.shape}')
print(f'Dimensiones del dataset de test: {test.shape}')
```

##### Guardamos los CSV train y test:

```python
train.to_csv('./train.csv', sep=';', decimal='.', index=False)
test.to_csv('./test.csv', sep=';', decimal='.', index=False)
```

##### Cargamos el dataset 'train.csv' con el que vamos a trabajar:

```python
airbnb_train = pd.read_csv('./train.csv', sep=';', decimal='.')
airbnb_train
```
